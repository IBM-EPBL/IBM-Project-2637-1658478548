## 6.Find and replace outliers

```python
sns.boxplot(data['Age'])
```

```python
data.mean()
```

```python
data.std()
```

```python
q = data.quantile(q=[0.75,0.5])
```

```python
iqr = q.iloc[0]-q.iloc[1]
```

```python
iqr
```

```python
l = q.iloc[1] - (1.5*iqr)
```

```python
l['Age']
```

```python
u = q.iloc[1] + (1.5*iqr)
```

```python
u['Age']
```

```python
sns.boxplot(data['Age'])
```

```python
data['Age'] = np.where(data['Age']>u['Age'],u['Age'],np.where(data['Age']<l['Age'],l['Age'],data['Age']))
```

```python
sns.boxplot(data['Age'])
```
