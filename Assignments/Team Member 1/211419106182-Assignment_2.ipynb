# Find the outliers and replace the outliers

```python
sns.boxplot(data['SOP'])
```

```python
data.mean()
```

```python
data.std()
```

```python
q = data.quantile(q=[0.85,0.3])
iqr = q.iloc[0]-q.iloc[1]
iqr
```

```python
l = q.iloc[1] - (1.5*iqr)
l['SOP']
```

```python
u = q.iloc[1] + (1.5*iqr)
u['SOP']
```

```python
sns.boxplot(data['SOP'])
```

```python
data['SOP'] = np.where(data['SOP']>u['SOP'],u['SOP'],np.where(data['SOP']<l['SOP'],l['SOP'],data['SOP']))
sns.boxplot(data['SOP'])
```
